# tempo zbieznosci metody ALS w zaleznosci od obranych parametrów

```python
colums = df.columns[1:]
greens = ['forestgreen', 'limegreen', 'chartreuse', 'lightgreen']
g=0
reds = ['red', 'tomato', 'salmon', 'lightcoral']
r=0
yellows=['yellow','gold','khaki','lemonchiffon']
y=0
blues=['darkblue','navy','royalblue','dodgerblue']
b=0
pinks=['fuchsia','m','orchid','violet']
p=0
```

```python
plt.figure(figsize=(8, 9), dpi= 80)
for name in colums:
    if name.startswith('d=3'):
        plt.plot(df['range'], df[name], color=greens[g])
        g +=1
    if name.startswith('d=5'):
        plt.plot(df['range'], df[name], color=reds[r])
        r +=1
    if name.startswith('d=8'):
        plt.plot(df['range'], df[name], color=yellows[y])
        y +=1
    if name.startswith('d=13'):
        plt.plot(df['range'], df[name], color=blues[b])
        b +=1
    if name.startswith('d=21'):
        plt.plot(df['range'], df[name], color=pinks[p])
        p +=1
plt.yscale('symlog')
plt.legend(loc='upper right')
plt.xlabel('Number Of Iterations')
plt.ylabel('Funkcja Celu')
plt.title('Zbieżnosc ALS dla macierzy 20x40')
plt.savefig('zbie1.png')
plt.show()
```

# Wpływ parametru d na jakość stworzonych rekomendacji i czas obliczeń

```python
import matplotlib.pyplot as plt

axis_x = (2, 3,5 ,8 ,13 ,21, 25, 28)
plt.figure(figsize=(16, 9), dpi= 80)
firstTime=[]
secondTime=[]
thirdTime=[]
for i in range(0, 8):
    firstTime.append(data_40_Users_20_Products[i]['time'])
    secondTime.append(data_400_Users_200_products[i]['time'])
    thirdTime.append(data_500_Users_1000_products[i]['time'])

plt.plot(axis_x, firstTime, color="black", label='Macierz 40x20')
plt.plot(axis_x, secondTime, color ="red", label='Macierz 400x200')
plt.plot(axis_x, thirdTime, color="purple",label='Macierz 1000x500')

    
plt.legend()
plt.yscale('symlog')
plt.xlabel('d')
plt.ylabel('czas(s)')
plt.title('Czas wykonywania metody ALS w zależności od parametru d dla lamda = 0.1')
plt.savefig('time.png')
plt.show()
```
